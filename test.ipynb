```python
def getCount(group_size, ratio):
    is_even = group_size % 2 == 0
    each_gend = group_size // 2
    high, low = max(ratio), min(ratio)
    
#When ratio is 1 : 1
    if high == low:
        if is_even == True:
            num_m = num_f = each_gend
        else:
            num_m = each_gend
            num_f = each_gend + 1      
    
#when the difference is equal to the group size
    elif high - low == group_size: 
        if high == ratio[0]:
            num_m = group_size
            num_f = 0
        else:
            num_m = 0
            num_f = group_size

#Difference less than or equal to 10
    elif high - low <= 10:
        if ratio[0] > ratio[1] and is_even:
            num_m = each_gend + 1
            num_f = each_gend - 1
        elif ratio[0] > ratio[1]:
            num_m = each_gend + 1
            num_f = each_gend
        elif ratio[0] < ratio[1] and is_even:
            num_m = each_gend - 1
            num_f = each_gend + 1
        else:
            num_m = each_gend
            num_f = each_gend + 1

#Difference larger than 10
    else:
        if high == ratio[0]:
            num_m = group_size
            num_f = 0
        else:
            num_m = 0
            num_f = group_size
    
    # print(num_m, num_f, end=":")
    return num_m,num_f
```

```python
def getSchlCount(list):
    schls_count = {}
    for schl, val in list.items():
        gends_count = {}
        total = 0
        for gend, v in val.items():
            # print(schl, ":", gend, len(v[0]))
            gends_count[gend] = len(v[0])
            total += len(v[0])
        gends_count["Total"] = total 
        schls_count[schl] = gends_count
    # print(schls_count)
    return(schls_count)
```

```python
def getSchlGend(list, gend, num_m, num_f):
    max_gend = []
    for sch, value in list:
        try:
            max_gend.append([sch, value[gend]])
        except:
            continue

    sorted_dict = sorted(max_gend, key=lambda x:x[1])

    group_size = num_m + num_f
    group = []

    while len(group) < group_size:
        if len(group) <= (group_size//2):
            schl = sorted_dict[len(group)][0]
        else:
            schl = sorted_dict[len(group)-group_size][0]
        group.append(schl)

    return group
```

```python
def getStudentbyGpaTest(list, avg_gpa, group_gpa):
    student = 0
    for i in list:
        ref_gpa = avg_gpa
        if group_gpa == avg_gpa:
            student = list.pop(list.index(i))
            break
        elif group_gpa > avg_gpa:
            if group_gpa - avg_gpa > 0.2:
                ref_gpa -= 0.2
            if i <= ref_gpa:
                student = list.pop(list.index(i))
                break
        else:
            if avg_gpa - group_gpa > 0.2:
                ref_gpa += 0.2
            if i >= ref_gpa:
                student = list.pop(list.index(i))
                break

    if student == 0:
        if group_gpa > avg_gpa:
            avg_gpa += 0.2
        else:
            avg_gpa -= 0.2
        student, list = getStudentbyGpaTest(list, avg_gpa, group_gpa)

    return student, list
```

```python
def getGenderCount(list):
    male = 0
    female = 0
    for v in list.values():
        #check gender ratio
        for gend, v in v.items():
            if gend == "Male":
                male += v[1]
            else:
                female += v[1]
    return((male, female))
```
